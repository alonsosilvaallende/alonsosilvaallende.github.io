TIL how to do quizzes with [JupyterQuiz](https://github.com/jmshea/jupyterquiz). JupyterQuiz is a tool to make quizzes in jupyter notebooks. It is very easy to use and you can even use it with WebAssembly as I will show below:

You can install `jupyterquiz` as follows:

::: {.panel-tabset}
## pip
```{.bash filename="Terminal"}
python -m venv .venv
source .venv/bin/activate
python -m pip install jupyterquiz
```
## uv
```{.bash filename="Terminal"}
uv venv
source .venv/bin/activate
uv pip install jupyterquiz
```
## wasm
```{.bash filename="Terminal"}
%pip install jupyterquiz
```
:::

Then you need to:

- Provide a question (in this example "What's the capital of France?"),
- Specify what's the type of question you want to make ("multiple_choice", "many_choice", etc.),
- and finally the possible answers and which one(s) are correct.

The expected format is as follows:

````python
#| echo: true
example = [
    {
        "question": "What is the capital of France?",
        "type": "multiple_choice",
        "answers": [
            {"answer": "London", "correct": False},
            {"answer": "Paris", "correct": True},
            {"answer": "New York", "correct": False},
            {"answer": "Rome", "correct": False},
        ],
    }
]
````

Then you can display the quiz:

````python
#| echo: true
from jupyterquiz import display_quiz

display_quiz(example)
````

You can also try it here by running the following cell:

````python
#| echo: false
code_str = """%pip install -q jupyterquiz

from jupyterquiz import display_quiz

example=[{
        "question": "What is the capital of France?",
        "type": "multiple_choice",
        "answers": [
            {
                "answer": "London",
                "correct": False
            },
            {
                "answer": "Paris",
                "correct": True
            },
            {
                "answer": "New York",
                "correct": False
            },
            {
                "answer": "Rome",
                "correct": False
            }
        ]
    }]

display_quiz(example)"""
````

````python
#| echo: false
import urllib.parse

code = urllib.parse.quote(code_str)
````

````python
#| echo: false
from IPython.display import IFrame

IFrame(f"https://jupyterlite.github.io/demo/repl/index.html?toolbar=1&kernel=python&promptCellPosition=left&hideCodeInput=1&clearCodeContentOnExecute=0&code={code}&execute=0", width=850, height=600)
````

You can also store the questions in a json file (in this case I'm storing it [here](https://github.com/alonsosilvaallende/alonsosilvaallende.github.io/blob/main/til/posts/2025-06-16-Jupyter_Quiz/examples.json)) and then display the quiz:

````python
#| echo: false
code_str = """%pip install -q jupyterquiz

from jupyterquiz import display_quiz

git_url='https://raw.githubusercontent.com/alonsosilvaallende/alonsosilvaallende.github.io/refs/heads/main/til/posts/2025-06-16-Jupyter_Quiz/examples.json'

display_quiz(git_url)"""
````

````python
#| echo: false
import urllib.parse

code = urllib.parse.quote(code_str)
````

````python
#| echo: false
from IPython.display import IFrame

IFrame(f"https://jupyterlite.github.io/demo/repl/index.html?toolbar=1&kernel=python&promptCellPosition=left&hideCodeInput=1&clearCodeContentOnExecute=0&code={code}&execute=0", width=850, height=400)
````

Notice that we are not storing the results so this is only for people to get feedback when using these quizzes.
